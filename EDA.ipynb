```python
# Titanic EDA - robust script (works with seaborn dataset or local CSV)
# Requirements: pandas, numpy, matplotlib, seaborn (only for loading dataset, optional)
# Run in Jupyter / Colab / local environment.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Try to load from seaborn (internet required). If not available, read local csv.
def load_titanic(fallback_path='titanic.csv'):
    try:
        import seaborn as sns
        df = sns.load_dataset('titanic')
        print("Loaded 'titanic' from seaborn.")
        return df
    except Exception as e:
        print("Could not load seaborn dataset (network may be blocked).")
        print("Attempting to load local CSV:", fallback_path)
        try:
            df = pd.read_csv(fallback_path)
            print("Loaded local CSV:", fallback_path)
            return df
        except Exception as e2:
            raise FileNotFoundError(
                f"Failed to load dataset. Provide a local CSV file at '{fallback_path}' "
                "or run in an environment with internet access."
            )

# Load dataset (adjust path if your CSV filename/path is different)
df = load_titanic('titanic.csv')

# Basic preview
print("\n--- Head of dataset ---")
display(df.head(10))   # in notebooks shows nicely; otherwise prints

# 1) Summary statistics
print("\n--- Numeric summary (describe) ---")
num_desc = df.describe()
print(num_desc)

print("\n--- Categorical summary (top / freq) ---")
cat_desc = df.describe(include=['object','category','bool'])
print(cat_desc)

# 2) Missing values summary
missing = df.isnull().sum().sort_values(ascending=False)
missing_pct = (df.isnull().mean()*100).sort_values(ascending=False)
missing_summary = pd.concat([missing, missing_pct], axis=1)
missing_summary.columns = ['missing_count', 'missing_percent']
print("\n--- Missing values (count, percent) ---")
print(missing_summary[missing_summary['missing_count']>0])

# 3) Histograms for numeric features (matplotlib used)
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print("\nNumeric columns detected:", numeric_cols)

for col in numeric_cols:
    plt.figure(figsize=(6,3.5))
    data = df[col].dropna()
    if data.empty:
        continue
    plt.hist(data, bins=20)
    plt.title(f'Histogram: {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.grid(axis='y', alpha=0.6)
    plt.tight_layout()
    plt.show()

# 4) Boxplots for numeric features grouped by survival (if 'survived' present)
if 'survived' in df.columns:
    for col in numeric_cols:
        if col == 'survived':
            continue
        df_box = df[[col, 'survived']].dropna()
        if df_box.empty:
            continue
        groups = [df_box[df_box['survived']==g][col] for g in sorted(df_box['survived'].unique())]
        plt.figure(figsize=(6,3.5))
        plt.boxplot(groups, labels=[f'survived={g}' for g in sorted(df_box['survived'].unique())])
        plt.title(f'Boxplot of {col} by survival')
        plt.ylabel(col)
        plt.tight_layout()
        plt.show()

# 5) Correlation matrix (numeric features)
corr = df[numeric_cols].corr()
print("\n--- Correlation matrix ---")
print(corr)

plt.figure(figsize=(6,5))
plt.imshow(corr, interpolation='nearest', aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha='right')
plt.yticks(range(len(corr.index)), corr.index)
plt.title('Correlation heatmap (numeric features)')
plt.tight_layout()
plt.show()

# 6) Scatter plots for selected pairs
pairs = []
if 'age' in df.columns and 'fare' in df.columns:
    pairs.append(('age','fare'))
if 'age' in df.columns and 'pclass' in df.columns:
    pairs.append(('age','pclass'))
if 'fare' in df.columns and 'pclass' in df.columns:
    pairs.append(('fare','pclass'))

for x,y in pairs:
    plt.figure(figsize=(6,4))
    sub = df[[x,y,'survived']].dropna() if 'survived' in df.columns else df[[x,y]].dropna()
    plt.scatter(sub[x], sub[y], s=12)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(f'{y} vs {x}')
    plt.grid(alpha=0.4)
    plt.tight_layout()
    plt.show()

# 7) Categorical counts and bar charts
cat_cols = df.select_dtypes(include=['object','category','bool']).columns.tolist()
print("\nCategorical columns:", cat_cols)
for col in cat_cols:
    counts = df[col].value_counts(dropna=False).head(12)
    print(f"\nValue counts for {col} (top):")
    print(counts)
    plt.figure(figsize=(6,3.5))
    counts.plot(kind='bar')
    plt.title(f'Value counts: {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()

# 8) Simple inferences
print("\n--- Basic inferences ---")
if 'survived' in df.columns:
    surv_rate = df['survived'].mean()
    print(f"Overall survival rate: {surv_rate:.3f} ({surv_rate*100:.1f}%)")
    if 'sex' in df.columns:
        print("\nSurvival rate by sex:")
        print(df.groupby('sex')['survived'].mean().sort_values(ascending=False))
    if 'pclass' in df.columns:
        print("\nSurvival rate by passenger class (pclass):")
        print(df.groupby('pclass')['survived'].mean().sort_values(ascending=False))

# 9) Save compact summary for later
summary_small = pd.DataFrame({
    'feature': df.columns,
    'dtype': df.dtypes.astype(str),
    'missing_count': df.isnull().sum(),
    'unique_count': df.nunique(dropna=False)
}).reset_index(drop=True)
out_path = 'titanic_summary.csv'
summary_small.to_csv(out_path, index=False)
print(f"\nSaved small summary to {out_path}")
```
